```python
df = pd.merge(df_mutation,df_patients,on='bcr_patient_barcode',how='left')
df.head()
```

```python
df['age_at_initial_pathologic_diagnosis'] = df['age_at_initial_pathologic_diagnosis'].replace('NaN',-1)
df['age_at_initial_pathologic_diagnosis'] = df['age_at_initial_pathologic_diagnosis'].replace(np.nan,-1)
df['age_at_initial_pathologic_diagnosis'] = df['age_at_initial_pathologic_diagnosis'].replace('[Not Available]',-1)

df['age_at_initial_pathologic_diagnosis'] = df['age_at_initial_pathologic_diagnosis'].astype('int64').astype('Int64')

df['age_at_initial_pathologic_diagnosis'] = df['age_at_initial_pathologic_diagnosis'].replace(-1,np.nan)

df.head()
```
